```python
import pandas as pd
import numpy as np
from datetime import date

data_path = "data"
output_path = "output/"
temp_path = "tmp/"

sample_submission = pd.read_csv(f"{data_path}/sample_submission.csv")
submission_cols = sample_submission.columns.tolist()
```

```python
# Load the test data
test_df = pd.read_parquet(f"{data_path}/test_data.parquet")
print(test_df.shape)
print(list(test_df.columns))
test_df.head()
```

```python
# Load the training data
train_df = pd.read_parquet(f"{data_path}/train_data.parquet")
print(train_df.shape)
print(list(train_df.columns))
train_df.head()
```

```python
cols = list(
    filter(
        lambda x: x.startswith("call") or x.startswith("put"), submission_cols
    )
)
cols
```

```python
test_df_mean = test_df.copy();
for col in cols:
    test_df_col = test_df_mean[col].copy()
    test_df_col.dropna(inplace=True)
    mean = test_df_col.mean()
    # print(mean)
    test_df_mean[col] = test_df_mean[col].fillna(mean)

print(test_df_mean.isna().sum().sum())
mean_submission = test_df_mean[submission_cols]
mean_submission.to_csv(f"{output_path}/test_mean_submission.csv", index=False)
```

```python
test_df_median = test_df.copy()
for col in cols:
    test_df_col = test_df_median[col].copy()
    test_df_col.dropna(inplace=True)
    median = test_df_col.median()
    # print(median)
    test_df_median[col] = test_df_median[col].fillna(median)

print(test_df_median.isna().sum().sum())
median_submission = test_df_median[submission_cols]
median_submission.to_csv(f"{output_path}/test_median_submission.csv", index=False)
```
